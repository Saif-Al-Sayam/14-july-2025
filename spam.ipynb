```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
import numpy as np

url = "https://raw.githubusercontent.com/Saif-Al-Sayam/14-july-2025/refs/heads/main/SMSSpamCollection"
df = pd.read_table(url, sep='\t', header=None, names=['label', 'message'])

df . head ( )
```

```python
# Step 2: Convert labels to binary values (ham=0, spam=1)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
df . head ( )
```

```python
# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], random_state=1)

print("Original dataset contains", df.shape[0], "messages")
print("Training set contains", X_train.shape[0], "messages")
print("Testing set contains", X_test.shape[0], "messages")
```

```python
# Step 4: Convert the text data into word counts (features)
count_vectorizer = CountVectorizer()
X_train_counts = count_vectorizer.fit_transform(X_train)
X_test_counts = count_vectorizer.transform(X_test)
# Step 5: Train a Naïve Bayes classifier
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train_counts, y_train)
# Step 6: Predict on the testing set
predictions = naive_bayes.predict(X_test_counts)
# Step 7: Evaluate the model accuracy
accuracy = np.mean(predictions == y_test)
print("Accuracy:", accuracy)
```

```python
# Step 8: Confusion matrix to evaluate performance on each class
conf_matrix = confusion_matrix(y_test, predictions)
print("Confusion Matrix:")
print(conf_matrix)
```
